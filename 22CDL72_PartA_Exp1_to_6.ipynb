## Experiment 4: Sampling Distribution of Mean (Central Limit Theorem)

```python
np.random.seed(42)
population = np.random.exponential(scale=5000, size=10000)
sample_means = [np.mean(np.random.choice(population, 50)) for _ in range(10)]
plt.hist(sample_means, bins=10, edgecolor='black')
plt.title('Distribution of Sample Means')
plt.xlabel('Mean Step Count')
plt.ylabel('Frequency')
plt.show()
print('Sample Means:', sample_means)
```

## Experiment 5: Weight of Evidence (WOE) Encoding â€“ Titanic Dataset

```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset (use your local path)
# Example: df = pd.read_csv("C:/Users/YourName/Downloads/titanic.csv")
df = pd.read_csv("C:/Users/naman/Downloads/Titanic-Dataset.csv")

# Step 2: Select relevant columns
df = df[['Survived', 'Sex', 'Pclass']]

# Step 3: Display the first few rows
print("First 5 rows of the dataset:")
print(df.head())

# Step 4: Define the WOE (Weight of Evidence) function
def woe(df, col, target):
    """
    Calculates Weight of Evidence (WOE) for each category in a feature.
    WOE = ln(Good% / Bad%)
    """
    lst = []
    for val in df[col].unique():
        # Count 'good' (target = 1) and 'bad' (target = 0)
        good = len(df[(df[col] == val) & (df[target] == 1)])
        bad = len(df[(df[col] == val) & (df[target] == 0)])
        
        # Compute distributions (add small value to avoid division by zero)
        dist_good = (good + 1e-6) / df[target].sum()
        dist_bad = (bad + 1e-6) / (len(df) - df[target].sum())
        
        # Calculate WOE
        woe_val = np.log(dist_good / dist_bad)
        lst.append([col, val, woe_val])
    
    return pd.DataFrame(lst, columns=['Variable', 'Value', 'WOE'])

# Step 5: Apply WOE encoding on 'Sex' column
woe_sex = woe(df, 'Sex', 'Survived')
print("\nWeight of Evidence for 'Sex':")
print(woe_sex)

# Step 6: Apply WOE encoding on 'Pclass' column
woe_pclass = woe(df, 'Pclass', 'Survived')
print("\nWeight of Evidence for 'Pclass':")
print(woe_pclass)

# Step 7: Combine WOE results
woe_all = pd.concat([woe_sex, woe_pclass])

# Step 8: Visualize the WOE values
plt.figure(figsize=(7,4))
sns.barplot(x='Value', y='WOE', hue='Variable', data=woe_all)
plt.title("Weight of Evidence (WOE) for Titanic Features")
plt.axhline(0, color='gray', linestyle='--')
plt.show()

print("\nWOE Encoding Completed Successfully.")
```

## Experiment 6: Bag of Words & TF-IDF (Text Encoding)

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
docs = [
    'Team wins championship after close match',
    'Player breaks world record in sprint',
    'Government passes new bill in parliament',
    'Elections to be held next month'
]
bow = CountVectorizer()
X_bow = bow.fit_transform(docs)
df_bow = pd.DataFrame(X_bow.toarray(), columns=bow.get_feature_names_out())
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(docs)
df_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=tfidf.get_feature_names_out())
print('Bag of Words:', df_bow)
print('TF-IDF:', df_tfidf)
```
